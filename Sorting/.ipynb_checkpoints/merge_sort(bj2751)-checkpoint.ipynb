# 병합정렬

이 코드는 백준 2751번을 병합 정렬로 푸는 구현입니다. 전체 배열 A를 시작 인덱스 s와 끝 인덱스 e로 구간을 나누어 재귀적으로 정렬하고, 각 구간을 병합할 때 tmp 배열에 현재 구간을 복사해 안전하게 비교합니다. 가운데 인덱스 m = s + (e - s) // 2를 기준으로 왼쪽과 오른쪽을 나누고, 두 포인터(index1, index2)를 움직이며 더 작은 값을 A[k]에 채워 넣습니다. 한쪽 구간이 먼저 소진되면 나머지 구간의 값들을 그대로 복사하여 전체 구간을 완성합니다. 이렇게 재귀와 병합을 반복해 전체 배열이 오름차순으로 정렬됩니다.

# Merge_sort

This code implements merge sort to solve Baekjoon problem 2751. It recursively sorts subarrays of the main array A, defined by a start index s and an end index e. Before merging, the current segment is copied into the auxiliary array tmp for safe comparison. The midpoint m = s + (e - s) // 2 is used to split the segment into left and right halves, and two pointers (index1, index2) are used to select the smaller value and place it into A[k]. When one side is exhausted, the remaining values from the other side are copied directly into A. Repeating this process of recursion and merging results in the entire array being sorted in ascending order.

```python
def merge_sort(s, e):
    if e - s < 1 : return
    m = int(s + (e - s) / 2)
    merge_sort(s,m)
    merge_sort(m + 1, e)
    for i in range(s, e + 1):
        tmp[i] = A[i]
    k = s
    index1 = s
    index2 = m + 1
    while index1 <= m and index2 <= e:
        if tmp[index1] > tmp[index2]:
            A[k] = tmp[index2]
            k = k + 1
            index2 = index2 + 1
        else:
            A[k] = tmp[index2]
            k = k + 1
            index1 = index1 + 1

    while index1 <= m:
        A[k] = tmp[index1]
        k = k + 1
        index1 = index1 + 1
    while index2 <= e:
        A[k] = tmp[index2]
        k = k + 1
        index2 = index2 + 1

N = int(input())
A = [0] * int(N + 1)
tmp = [0] * int(N + 1)

for i in range(1, N + 1):
    A[i] = int(input())

merge_sort(1, N)

for i in range(1, N+1):
    print(str(A[i]) + '\n')
```
