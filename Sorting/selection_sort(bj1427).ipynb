# 선택정렬
# Selection Sort

선택정렬이란 대상 데이터에서 **최대**나 **최소** 데이터를 데이터가 나열된 순으로 찾아가며 선택하는 방법이다.

선택 정렬 과정

1. 남은 정렬 부분에서 최솟값 또는 최댓값을 찾는다
2. 남은 정렬 부분에서 **가장 앞에 있는 데이터**와 **선택된 데이터**를 swap한다.
3. 가장 앞에 있는 데이터의 위치를 변경해(index++) 남은 정렬 부분의 범위를 축소한다.
4. 전체 데이터 크기만큼 index가 커질 때까지, 즉 남은 정렬 부분이 없을 때까지 반복한다.

#### 핵심개념: 이번 차례(i)에서 남은 사람 중 제일 큰 값을 앞으로 swap해서 정렬을 완성하는 과정!
i번째 자리(현재 차례)에 있는 값이,

그 뒤에 남아있는 값들 중(j번째) 더 큰 값이 있으면,

**둘을 자리 바꿔준다(swap)**는 뜻이야.

Selection sort is a method that finds the maximum or minimum value from the unsorted data and selects it in order.

Selection Sort Process
Find the minimum or maximum value from the unsorted portion of the data.

Swap the value at the beginning of the unsorted portion with the selected value.

Move the index forward (index++), reducing the range of the unsorted portion.

Repeat this process until the index reaches the end of the data, meaning there is no unsorted portion left.

Key concept:
At each step (i), you search for the largest (or smallest) value remaining and swap it with the value at the current position.
If there is a value among the remaining elements (j) that is bigger than the current value (i), you swap them.




```python
# 파이썬에서는 input 데이터를 list로 변환하면 각 자릿수로 나누어 리스트화해주기 때문에 변경이 매우 간편하다.
A = list(input())

for i in range(len(A)):
    Max = i
    for j in range(i+1, len(A)): #i 뒤부터 끝까지 중에서 가장 큰 값을 찾는 for문이야.
        if A[j] > A[Max]: # 만약 j번째 값이 지금까지 찾은 가장 큰 값보다 크다면 가장 큰 값의 위치(Max)를 j로 갱신!
            Max = j # 이번 회차에서 가장 큰 값의 인덱스를 찾으려고 시작값을 i로 지정함.
    if A[i] < A[Max]: # 만약, i번째 값이 이번 회차에서 찾은 가장 큰 값(Max)보다 작다면 두 값을 자리 바꿈(swap) 해줌
        temp = A[i]
        A[i] = A[Max]
        A[Max] = temp

for i in range(len(A)):
    print(A[i])
```
