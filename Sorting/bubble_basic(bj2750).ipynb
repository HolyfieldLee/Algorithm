### 버블정렬은 이중 for문을 이용해 인접한 값들을 temp 변수로 자리를 바꿔가며 정렬하는 방법이다.
### English : Bubble sort is a method of sorting adjacent values by swapping their positions using a temp variable using a double for loop

* 왜 range(N-1)=> 배열의 모든 숫자를 정렬하려면, 최대 N-1번 반복하면 돼.(N번째에는 이미 다 정렬되어 있으니까!)
* 왜 range(N-1-i)인가? => 안쪽 for문에서는 현재 남아있는 정렬되지 않은 구간만큼만 반복하면 돼.
* 왜냐하면, 한 번 반복할 때마다 맨 끝에 가장 큰 값이 확정되기 때문이야!

1회차 끝나면, 제일 큰 값이 맨 뒤로 감

2회차 끝나면, 두 번째로 큰 값이 그 앞에 감

... 이런 식으로

그래서 매 반복마다 비교해야 하는 구간이 1씩 줄어들어!

첫 바퀴: N-1번 비교

두 번째 바퀴: N-2번 비교

세 번째 바퀴: N-3번 비교

...

이걸 코드로 쓰면 range(N-1-i)가 되는 거야.

##### 한 줄 요약
* 바깥 for문: 전체를 N-1번 반복하며 큰 수가 뒤로 갈 수 있도록 함.

* 안쪽 for문: 이미 뒤로 간 큰 수는 더 이상 안 건드리도록, 남은 부분만 비교해서 정렬함.

English:

* Why range(N-1)?
→ To sort all the numbers in the array, you only need to repeat the process up to N-1 times. By the Nth pass, the array is already sorted!

* Why range(N-1-i)?
→ In the inner for loop, you only need to check the unsorted portion of the array each time.

* Why?
→ Because after each pass, the largest number is guaranteed to be at the end of the array!

After the 1st pass, the largest number moves to the last position.

After the 2nd pass, the second largest number moves just before the last position.

... and so on.

That’s why, with each pass, the range you need to compare gets one smaller.

1st pass: N-1 comparisons

2nd pass: N-2 comparisons

3rd pass: N-3 comparisons

...

In code, this is written as range(N-1-i).

##### One-line summary
* The outer loop: Repeats N-1 times so that larger numbers can bubble up to the end of the array.

* The inner loop: Only compares the unsorted part, skipping numbers that have already been placed at the end.



```python
N = int(input())
A = [0]*N

for i in range(N):
    A[i] = int(input())

for i in range(N-1):
    for j in range(N-i-1):
        if A[j] > A[j+1]:
            temp = A[j]    # 크다고 조건되어 있는 것을 temp로 담자! 이렇게 암기하면 됨
            A[j] = A[j+1]  # English: Let's store the thing that is conditioned to be large as temp! Just memorize it like this.
            A[j+1] = temp
            
for i in range(N):
    print(A[i])

```
